```python
from tkinter import *
from tkinter import messagebox


def ventana_login():
    raiz = Tk()
    raiz.title("Login Sortija")
    raiz.resizable(False, False)
    raiz.iconbitmap('Flechas.ico')
    raiz.config(bg="blue")
    
    marco = Frame(raiz)
    marco.config(width=300, height=130, bg="green")
    marco.pack()
    
    nombre = StringVar()
    clave = StringVar()

    #Casilla de Usuario
    mensaje1 = Label(marco, text="Usuario Alumno: ")
    mensaje1.grid(row=0, column=0, sticky= "e", padx=10, pady=10)

    entry1 = Entry(marco, textvariable=nombre)
    entry1.grid(row=0, column=1, padx=10, pady=10)
    
    #Casilla de clave
    mensaje2 = Label(marco, text="Clave: ")
    mensaje2.grid(row=1, column=0, sticky= "e", padx=10, pady=10)

    entry2 = Entry(marco, show="*", textvariable=clave)
    entry2.grid(row=1, column=1, padx=10, pady=10)
    
    def validar_clave():
        nombre_usuario = nombre.get()
        clave_usuario = clave.get()
        claves = obtener_usuarios_claves()
        control = False
        if nombre_usuario in claves.keys():
            if claves[nombre_usuario] == clave_usuario:
                control = True
                mostrar_mensaje(control)
            else:
                mostrar_mensaje(control)
        else:
            mostrar_mensaje(control)  
    
    #Boton enviar formulario
    nombre_usuario = nombre.get()
    clave_usuario = clave.get()
    print(nombre_usuario, clave_usuario)
    botonEnvio=Button(marco, text="Enviar Formulario", command=validar_clave)
    botonEnvio.grid(row=2, column=0, padx=10, pady=10)
    
    raiz.mainloop()
    
def obtener_usuarios_claves():
    usuarios_claves={
        "German": "key123",
        "Camila": "pro571",
        "Daniel": "reg708",
        "Federico": "les446",
        "Elias": "wot798",
    }
    return usuarios_claves

def mostrar_mensaje(control):
    if control: 
        messagebox.showinfo(title=None, message="Usuario y Clave Correctos")
    else:
        messagebox.showerror(title=None, message="Algunos de los datos ingresados es Incorrecto")
               
        
    
def validar_clave(nombre_usuario, clave_usuario):
        claves = obtener_usuarios_claves()
        if nombre_usuario in claves.keys():
            if claves[nombre_usuario] == clave_usuario:
                mostrar_info("Usuario y Clave Correctos")
            else:
                mostrar_error("Algunos de los datos ingresados es Incorrecto")
        else:
            mostrar_error("Algunos de los datos ingresados es Incorrecto")


ventana_login()
```
